```python
import pandas as pd
from datetime import *
```

```python
df = pd.read_csv("complaints.csv")
df.head()
```

```python
df = df.assign(year = 0)

for i in df.index:
    df.at[i, 'year'] = datetime.strptime(df[i, 'Date received'], '%Y-%m-%d').year
```

```python
df = df[ df['year'] == '2017' ]
df = df[ df['Consumer complaint narrative'] != 'NA' ]
```
